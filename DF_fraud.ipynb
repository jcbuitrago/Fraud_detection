##Cleaning Dataframe

Counting number of empty rows per column. To check for useless features.

```python
empty_columns = df.isna().sum()
```

```python
empty_columns.sort_values()
```

Defining a limit of empty rows on a column for it to be eliminated.

```python
cols_to_delete = empty_columns[empty_columns >= int(len(df)*0.1)].index
len(cols_to_delete)
```

```python
clean_df = df.drop(columns=cols_to_delete)
clean_df.shape
```

```python
isFraud_df = clean_df.loc[clean_df['isFraud'] == 1]
notFraud_df = clean_df.loc[clean_df['isFraud'] == 0]
print('isFraud: ' + str(isFraud_df.shape) + '     notFraud: ' + str(notFraud_df.shape))
```
